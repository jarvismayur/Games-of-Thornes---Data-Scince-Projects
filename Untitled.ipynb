```python
# Creating  a actor dialogue maps for every character from the text files. 

from collections import defaultdict

def get_actor_dialog_map(filename:str)->dict:
  with open(filename) as file:
    lines = file.readlines()
    lines = [line.strip() for line in lines if line.strip() != ''] # Make sure to remove empty lines
  # Make tuple of actor and dialog.
  tuple_ls = [(i.split(':')[0],i.split(':',maxsplit=2)[1] )for i in lines]
  actor_dialog_map = defaultdict(list)
  # Optimized this 
  for actor,dialog in tuple_ls:
    actor_dialog_map[actor].append(dialog.strip())
  return actor_dialog_map
```
